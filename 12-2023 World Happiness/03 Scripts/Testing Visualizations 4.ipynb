# Merge Main Dataframe with New Dataframe in Preparation of Time Series Analysis

```python
main_df.head()
```

```python
df.head()
```

```python
# Melt the DataFrame to reshape it
df_melted = pd.melt(df, id_vars=['Country Name'], var_name='Year', value_name='Actual_GDP_Per_Capita')

# Print the updated df_melted DataFrame
print(df_melted.head())
```

```python
df_melted.dtypes
```

```python
main_df.dtypes
```

```python
# Convert the "Year" column in df_melted to int64
df_melted['Year'] = df_melted['Year'].astype(int)

# Merge main_df with df_melted using a left merge
merged_df = pd.merge(main_df, df_melted, how='left', left_on=['Country', 'Year'], right_on=['Country Name', 'Year'])

# Drop the duplicate "Country Name" column and unnecessary columns
merged_df.drop(['Country Name'], axis=1, inplace=True)

# Print the updated merged_df
merged_df.head()
```

```python
# Check for null values in the DataFrame df
null_values = merged_df[merged_df.isnull().any(axis=1)]

# Print the rows with null values
null_values
```

# Drop Taiwan, Northern Cyprus, Somaliland, South Sudan from our data due to not having GDP available.

```python
# This will drop all rows that contain at least one null value
merged_df = merged_df.dropna()

# Check for null values in the DataFrame df
null_values = merged_df[merged_df.isnull().any(axis=1)]

# Print the rows with null values
null_values
```
