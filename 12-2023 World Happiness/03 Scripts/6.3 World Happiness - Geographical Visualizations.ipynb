```python
import folium
import json

# GeoJSON file path
geojson_path = r'C:\Users\natha\OneDrive\Desktop\Data Analytics\Jupyter\12-2023 World Happiness\02 Data\Geospatial Libraries\countries.geo.json'

# Load the GeoJSON file using json module
with open(geojson_path, 'r') as file:
    world_geojson = json.load(file)

# Get the unique country names from the GeoJSON file
unique_countries_geojson = set([feature['properties']['name'] for feature in world_geojson['features']])

# Print the list of unique countries in the GeoJSON file
print("Unique Countries in GeoJSON:", unique_countries_geojson)
```

```python
# Get the unique country names from the DataFrame
unique_countries_df = set(df['Country'])

# Print the list of unique countries in the DataFrame
print("Unique Countries in DataFrame:", unique_countries_df)
```

```python
# Find discrepancies
df_not_in_geojson = unique_countries_df - unique_countries_geojson

# Print the names of countries from your DataFrame that don't match the GeoJSON file
print("Countries in DataFrame but not in GeoJSON:", df_not_in_geojson)
```

```python
from fuzzywuzzy import process

# Function to find the best match for each country
def find_best_match(country, choices):
    return process.extractOne(country, choices)

# Dictionary to store the corrected country names
corrections = {}

# Iterate through countries in DataFrame not in GeoJSON
for country in df_not_in_geojson:
    # Find the best match from the GeoJSON countries
    match, score = find_best_match(country, unique_countries_geojson)
    
    # Add the correction to the dictionary
    corrections[country] = match

# Print the corrections
print("Corrections:", corrections)
```

```python
corrections = {
    'Palestine': 'West Bank',
    "Côte d'Ivoire": 'Ivory Coast',
    'Czechia': 'Czech Republic',
    'eSwatini': 'Swaziland',
    'Congo': 'Republic of the Congo',
    'S. Sudan': 'South Sudan',
    'N. Cyprus': 'Northern Cyprus',
    'Dem. Rep. Congo': 'Democratic Republic of the Congo',
    'Dominican Rep.': 'Dominican Republic',
    'Serbia': 'Republic of Serbia',
    'Tanzania': 'United Republic of Tanzania',
    'Central African Rep.': 'Central African Republic',
    'Bosnia and Herz.': 'Bosnia and Herzegovina'
}

# Apply corrections to the 'Country' column in your DataFrame
df['Country'] = df['Country'].replace(corrections)

# Print the updated DataFrame
print(df)
```

## With newly imputed Data, we may have an incorrect "Happiness_Rank", we can re-value this rank easily because it is directly correlated to the "Happiness_Score"

```python
# Sort the DataFrame by Year and Happiness_Score in descending order
df.sort_values(by=['Year', 'Happiness_Score'], ascending=[True, False], inplace=True)

# Reset the index to reflect the new order
df.reset_index(drop=True, inplace=True)

# Create a new column 'New_Happiness_Rank' with the recalculated rank
df['New_Happiness_Rank'] = df.groupby('Year').cumcount() + 1

# Drop the old 'Happiness_Rank' column and rename the new one
df.drop(columns=['Happiness_Rank'], inplace=True)
df.rename(columns={'New_Happiness_Rank': 'Happiness_Rank'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Region and calculate the average Happiness Score
region_avg_happiness = df.groupby('Region')['Happiness_Score'].mean().sort_values(ascending=False)

# Plotting the results
plt.figure(figsize=(12, 8))
sns.barplot(x=region_avg_happiness.values, y=region_avg_happiness.index, palette='viridis_r')
plt.title('Average Happiness Score by Region')
plt.xlabel('Average Happiness Score')
plt.ylabel('Region')
plt.show()
```
